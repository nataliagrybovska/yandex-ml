```python
import numpy as np
from matplotlib import pylab as plt
import math
from numpy import arange, sin, exp
%matplotlib inline
```

```python
def f(x): 
    return sin(x / 5.) * exp(x / 10.) + 5 * exp(-x / 2.)
```

```python
_x = np.arange(0, 16, 0.1)
plt.plot(_x, f(_x))



```

```python

x1 = 1
x2 = 4
x3 = 10
x4 = 15

a = np.array([[1, x1, x1**2, x1**3], [1, x2, x2**2, x2**3], [1, x3, x3**2, x3**3], [1, x4, x4**2, x4**3]])
b = np.array([f(x1), f(x2), f(x3), f(x4)])
x = np.linalg.solve(a, b)

def p(_x):
    return x[0] + x[1]*_x  + x[2]*_x**2 + x[3]*_x**3

_x = np.arange(0, 16, 0.1)
plt.plot(_x, p(_x))
plt.plot(_x, f(_x))
```

```python
result_line = " ".join(map(str, x))

submission_file = open('submission2.txt', 'w')
submission_file.write(result_line)

submission_file.close()
```
