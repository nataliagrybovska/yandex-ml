```python
from sklearn import datasets
from sklearn import ensemble, cross_validation, tree
%pylab inline
```

```python
def write_answer(name, answer):
    with open(name + ".txt", "w") as fout:
        fout.write(str(answer))
```

```python
digits = datasets.load_digits()
```

```python
X = digits.data
y = digits.target

n_features = X.shape[1]
```

```python
from sklearn.model_selection import cross_val_score
tree_classifier = tree.DecisionTreeClassifier()

cv_score = cross_val_score(tree_classifier, X, y, cv=10)
answer1 = cv_score.mean()
print answer1
write_answer("1", answer1)
```

```python
bagging_classifier = ensemble.BaggingClassifier(base_estimator=tree_classifier, n_estimators=100)

answer2 = cross_val_score(bagging_classifier, X, y, cv=10).mean()
print answer2
write_answer("2", answer2)
```

```python
max_features= int(n_features**(1/2.))

bagging_classifier2 = ensemble.BaggingClassifier(base_estimator = tree_classifier, n_estimators = 100, max_features = max_features)

answer3 = cross_val_score(bagging_classifier2, X, y, cv=10).mean()
print answer3
write_answer("3", answer3)
```

```python
bagging_classifier3 = ensemble.BaggingClassifier(base_estimator = tree.DecisionTreeClassifier(max_features = max_features), n_estimators = 100)

answer4 = cross_val_score(bagging_classifier3, X, y, cv=10).mean()
print answer4
write_answer("4", answer4)
```

```python
rf_classifier = ensemble.RandomForestClassifier(n_estimators = 100, max_features = max_features)


answer5 = cross_val_score(rf_classifier, X, y, cv=10).mean()
print answer5
```
